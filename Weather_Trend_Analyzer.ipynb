```python
import random
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Sample weather data (project demo ke liye)
temperature = [30, 32, 31, 34, 35, 33, 36]
humidity = [60, 58, 65, 62, 59, 61, 57]

print("Weekly Weather Report\n")

for i in range(len(days)):
    print(days[i], "-> Temp:", temperature[i], "°C | Humidity:", humidity[i], "%")

# Prediction
if temperature[-1] > temperature[-2]:
    print("\nPrediction: Temperature may rise tomorrow.")
else:
    print("\nPrediction: Temperature may fall tomorrow.")

# Graph
plt.plot(days, temperature, marker='o', label="Temperature")
plt.plot(days, humidity, marker='o', label="Humidity")

plt.title("Weekly Climate Pattern")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```

Introduction:

Weather Trend Analyzer is a Python-based project designed to monitor and analyze weather conditions over a period of time. It helps in understanding changes in temperature and humidity through graphical representation.

Objective:

The main objective of this project is to track weather patterns, compare weekly climate data, and predict future temperature trends using Python.

Methodology:

The project collects weather-related data such as temperature and humidity, stores it in Python lists, and processes it for analysis. It then uses graphical visualization to represent the changing weather conditions.

Result:

The system successfully displays weekly weather reports, identifies climate trends, and predicts whether the temperature is likely to rise or fall.

Conclusion:

This project provides a simple way to analyze weather patterns and visualize climate changes, making it easier to understand weather behavior over time.
